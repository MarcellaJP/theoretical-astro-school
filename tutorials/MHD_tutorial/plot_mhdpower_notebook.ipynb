Demo power spectral density on a MHD time series


```python
import numpy as np
from matplotlib import pyplot as plt
from astropy.io import fits
from scipy import signal
```

Read in the data


```python
#Load in the data image using astrophy fits loader

image_data = fits.getdata('./TurbSignal.fits', ext=0)
print(image_data.shape)

sig = image_data


print(sig.shape)
f=np.arange(256)

# now that you loaded in the data and made an array of matching frequency,compute the power spectrum (use google, scipy is a clue!)


```

```python
freqs, times, spectrogram = signal.spectrogram(sig)
plt.figure(figsize=(5, 4))
plt.imshow(spectrogram, aspect='auto', origin='lower') #, cmap='hot_r', origin='lower')
plt.title('Spectrogram')
plt.ylabel('Frequency band')
plt.xlabel('Time window')
plt.tight_layout()
```

```python
freqs, psd = signal.welch(sig)

plt.figure(figsize=(5, 4))
plt.semilogx(freqs, psd)
plt.title('PSD: power spectral density')
plt.xlabel('Frequency')
plt.ylabel('Power')
plt.tight_layout()
```
